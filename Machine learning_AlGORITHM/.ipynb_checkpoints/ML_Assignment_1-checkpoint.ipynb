**Q1**. What is gradient descent algorithm? What is its function?

Gradient descent is an optimization algorithm used to minimize some function by iteratively moving in the direction of steepest descent as defined by the negative of the gradient. In machine learning, we use gradient descent to update the parameters of our model

**Q2.** Teach your computer to approximate the given function using linear regression and gradient descent. 

f(x) = 17x + 29

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# generate random data-set
np.random.seed(0)
x = np.random.rand(100, 1)
y = 29 + 17 * x + np.random.rand(100, 1)

# sckit-learn implementation

# Model initialization
regression_model = LinearRegression()
# Fit the data(train the model)
regression_model.fit(x, y)
# Predict
y_predicted = regression_model.predict(x)

# model evaluation
mse = mean_squared_error(y, y_predicted)
r2 = r2_score(y, y_predicted)

# printing values
print('Coefficients:' ,regression_model.coef_)
print('Intercept:', regression_model.intercept_)
print('mean squared error: ', mse)
print('R2 score: ', r2)

# plotting values

# data points
plt.scatter(x, y, s=10)
plt.xlabel('x')
plt.ylabel('y')

# predicted values
plt.plot(x, y_predicted, color='r')
plt.show()
```

```python
# Import the necessary libraries
import numpy
import matplotlib.pyplot as plot
import pandas
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Import the dataset into pandas dataframe
dataset = pandas.read_csv('housing.csv')
RM      = dataset.iloc[:, -1:].values
LSTAT   = dataset.iloc[:, :1].values
PTRATIO = dataset.iloc[:, -1:].values
MEDV    = dataset.iloc[:, :1].values
dataset.head(10)
```

```python
dataset.shape
```

```python
# Split the dataset into the training set and test set

RMTrain, RMTest, LSTATTrain,LSTATTest,PTRATIOTrain,PTRATIOTest,MEDVTrain, MEDVTest= train_test_split(RM,LSTAT,PTRATIO,MEDV, test_size = 0.33, random_state = 5)
```

```python
print(RMTrain.size)
print(RMTest.size)
print(LSTATTrain.size)
print(LSTATTest.size)
print(PTRATIOTrain.size)
print(PTRATIOTest.size)
```

```python
RMTrain
```

```python
# Creating a LinearRegression object and fitting it on our training set.
linearRegressor = LinearRegression()
linearRegressor.fit(RMTrain, LSTATTrain)
linearRegressor.fit(PTRATIOTrain, MEDVTrain)

# Predicting the test set results
RMPrediction  = linearRegressor.predict(RMTest)
LSTATPrediction = linearRegressor.predict(LSTATTest)
PTRATIOPrediction = linearRegressor.predict(PTRATIOTest)
MEDVPrediction = linearRegressor.predict(MEDVTest)
```

```python
print(RMPrediction)
print(LSTATPrediction)
print(PTRATIOPrediction)
print(MEDVPrediction)
```

```python
print(RMTest)
print(LSTATTest)
print(PTRATIOTest)
print(MEDVTest)
```

```python
print('Coefficients: \n', linearRegressor.coef_)
print('Intercept: \n', linearRegressor.intercept_)
```

```python
# Visualising the training set results
plot.scatter(RMTrain,LSTATTrain, color ='red')
plot.scatter(PTRATIOTrain,MEDVTrain,c='r')
plot.plot(RMTrain  ,linearRegressor.predict(RMTrain),color= 'blue')
plot.plot(LSTATTrain,linearRegressor.predict(LSTATTrain),color= 'red')
plot.plot(PTRATIOTrain,linearRegressor.predict(PTRATIOTrain),color= 'green')
plot.plot(MEDVTrain,linearRegressor.predict(MEDVTrain),color='blue')
plot.title('Housing vs Experience (Training set)')
plot.xlabel('Experience')
plot.ylabel('Price')
plot.show()
```

```python
# Visualising the test set results
plot.scatter(RMTest, LSTATTest, color = 'red')
plot.scatter(PTRATIOTest,MEDVTest,c='r')
plot.plot(RMTrain, linearRegressor.predict(RMTrain), color = 'blue')
plot.title('Housing vs Experience (Test set)')
plot.xlabel('Housing vs Experience')
plot.ylabel('price')
plot.show()
```
